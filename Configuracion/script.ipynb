# [dbo].[CRUD_CLIENTE]

```sql
/****** Object:  StoredProcedure [dbo].[CRUD_VENDEDOR]    Script Date: 13/11/2022 11:47:57 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[CRUD_VENDEDOR]

@busqueda varchar (50) = '',
@id int = 0, 
@Nombre varchar (50) = '', 
@Apellido varchar (50) = '',
@Telefono varchar (50) = '',
@Edad int = 0,
@Caja int = 0, 
@opcion char (1)


as 
   IF @opcion = 'I'
   begin 
        INSERT INTO Vendedor (id, Nombre, Apellido, telefono, edad, caja)
		                values(@id, @Nombre, @Apellido, @Telefono, @Edad,@Caja);

						declare @idVendedor as int = (select max (id) from Vendedor)

						insert into Pedido (idVendedor)
						values (@idVendedor)




   end 

   IF @opcion  = 'U'
   BEGIN 
         UPDATE Vendedor set id =@id,
		                     Nombre = @Nombre,
							 Apellido = @Apellido,
							 telefono = @Telefono, 
							 Edad = @Edad,
							 caja = @caja
		 where id = @id OR Nombre = @Nombre

	END 


	IF @opcion = 'D'
	BEGIN 
	     DELETE FROM Vendedor

		 WHERE id = @id or Nombre = @Nombre


		 DELETE FROM Pedido

		 WHERE id = @id

	end 

	IF @opcion = 'M'

	BEGIN 
	     SELECT id,
		        Nombre,
				Apellido,
				Telefono,
				edad, 
				caja
		 FROM Vendedor 
		 WHERE id like '%' + @busqueda + '%' or Nombre like '%' + @busqueda + '%' or Apellido like '%' +@busqueda + '%'

	end 
GO
```
